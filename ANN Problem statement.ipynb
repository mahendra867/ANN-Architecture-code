```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
df = pd.read_csv("C:\\Users\\mahen\\Downloads\\Churn_Modelling\\Churn_Modelling.csv")
df.head()
```

```python
# dividing the dataset into Independent and dependent features
x=df.iloc[:,3:13]
y=df.iloc[:,13]
```

```python
print(f"The independent features{x.head()} are")
```

```python
print(f"The Target feature is {y.head()} ")
```

```python
## Handling categorical features
#Create dummy variables by one hot encoding
geography=pd.get_dummies(x["Geography"],drop_first=True)
gender=pd.get_dummies(x['Gender'],drop_first=True)
```

```python
## Concatenate the Data Frames

x=pd.concat([x,geography,gender],axis=1)

## Drop Unnecessary columns
x=x.drop(['Geography','Gender'],axis=1)

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) # to avoid the data leakage we fit transform on train and only transform in test data
X_test = sc.transform(X_test)
```

## what is tensorflow and keras 
What is TensorFlow and Keras?
TensorFlow:

- TensorFlow is an open-source deep learning framework developed by the Google Brain team. It provides a comprehensive ecosystem of tools, libraries, and community resources that allows researchers and developers to build and deploy machine learning models easily.
- TensorFlow is designed for large-scale machine learning tasks, including deep neural networks, and it can run on multiple CPUs, GPUs, and TPUs (Tensor Processing Units).
Keras:

- Keras is a high-level neural networks API that simplifies the process of building and training deep learning models. It provides a user-friendly interface with simple, concise functions and clear error messages.
Originally, Keras was a standalone library that could run on top of several backend engines, including TensorFlow, Theano, and Microsoft Cognitive Toolkit (CNTK). However, with the release of TensorFlow 2.0, Keras is now tightly integrated with TensorFlow as the high-level API, which means it is included as part of the TensorFlow library.

### Relationship Between TensorFlow and Keras:
- Keras as a Wrapper: Keras acts as a wrapper for TensorFlow, allowing users to define and train neural networks using simple and intuitive commands. This abstraction helps users avoid the complexities of directly working with TensorFlow's lower-level operations.
- Integration in TensorFlow 2.0 and Above: In TensorFlow 2.0 and later versions, Keras is fully integrated, meaning you can access it directly through the tf.keras module. This integration makes it easier to develop models with TensorFlow while benefiting from Keras' simplicity.


```python
from tensorflow.keras.models import Sequential # Usaully we know Neural network architecture consist input,hidden,output layers so with help of  Sequential we can perform the Forward and Backward propagation in the Neural network  
from tensorflow.keras.layers import Dense  # With the help of Dense we can able to create the Neurons (circles) in the I/p layer , hidden layer and output layer  
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU # with the helps of activation fucntions we are going to use in the different layers so that it helps us to determine whether the neuron got activated or not
from tensorflow.keras.layers import Dropout #Dropout is a regularization technique used to reduce overfitting in neural networks.  Problem: Overfitting occurs when a model performs well on training data but poorly on test data, indicating it has memorized the training data instead of generalizing from it. Mechanism: During training, dropout randomly deactivates a specified percentage of neurons in a layer. For example, a dropout rate of 0.3 means 30% of the neurons in that layer will be deactivated.Effect: By deactivating neurons, dropout prevents the network from becoming overly reliant on specific neurons, forcing it to learn more robust features that generalize better to new data.Implementation: The dropout layer is added to the network during training, and neurons are randomly deactivated during each forward and backward propagation step. Deactivated neurons have their connections cut off, effectively making them inactive.
```

```python
 # Initlizing the ANN
classifier=Sequential() # when we decide the entier ANN is going to iniltize in the form of sequencial which consist of i/p layer, hidden layer, output layer and  which iam going to store in classifier variable 
```

```python
# Adding the input layer
classifier.add(Dense(units=11,activation='relu')) # here i have added input layer with the help of Dense which helps us to add neurons in the input layer in such a way that i haved added the neurons which are total 11 and which i mentioned in the form of units and when untis =11 means we add neurons in the input layer and i have mentioned the Relu activation fucntion which this activation function is going to utilize or apply in the next layer which is hidden layer and i have give units =11 because in the train shape i got 11 independent features to get train by this ANN that why i created 11 neurons to get pass this 11 features data further
```

```python
# Adding first Hidden layer
classifier.add(Dense(units=7,activation='relu')) # and in this first hidedn layer i have take n the 7 neurons and
classifier.add(Dropout(0.2))  # her i have added the dropout in the first hidden layer telling my model to drop 20% neurons or cut the connection of 20% of neurons during training 
```

```python
# Adding the second hidden layer 
classifier.add(Dense(units=6,activation='relu'))
classifier.add(Dropout(0.3))
```

```python
# Adding the output layer
classifier.add(Dense(units=1,activation='sigmoid')) # since we are dealing with classification problem statment we need to use sigmoid activation function in the outlayer in order to preidct either 0 or 1 and i have given the 1 neuron in the output layer to get prediction 0 or 1 
```

#### By the above lines of code i succesfully created the ANN architecture so  now i need to train the ANN model which i will write the code for do train the model  in the below lines of block

```python
# Now the Training the ANN model
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) # Training the model is build by compile function which first i need top mention the optimizer , loss, and then metircs and i have used the adam optimizer because its is the best optimizer in the present, loss function for binary classification problme is binary_crossentropy and metrics i have used is accuracy and by default the learning rate applied for taken optimizer is 0.01 while training the model
```

```python
# mentioining the customize learning rate parameter value for the adam optimizer by belopw code
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01) # here i have given the customized learning rate which is 0.02
```

```python
# Finally performing the training by fit method without early stopping parameter
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=50) # here i have given the validation set as 33% because my out of 100% (actual 80% of train data ) of train data which we given the 33% for validation set for calculating the loss we need to give the validation set which means loss function to find the minimal distance between the actual target feature values  and predicted values and we get predicted values by this validation set , and batch size we given as 10 which means at a time 10 records are going to feed to model and get perform the F+B propagation to miinze the loss function by using opitimizer and we given max Forward+backward propagation for each batch size(10) as 1000  for updating the weights
```

```python
model_history.history.keys() # as we can see here below accuracy, loss for the training set and val_accuracy, val_loss for validation set
```

### importance of early stopping 

```python
# Early stopping = it make sures that when accuracy is not at all increasing or when the validation loss is not getting imporved while updating the weights of batch of records thorugh no of epochs , the model training will stop  automatically
import tensorflow as tf
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=0,
    verbose=0,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
# Finally performing the training by fit method with early stopping parameter
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=50,callbacks=early_stopping) # here i have given the validation set as 33% because my out of 100% (actual 80% of train data ) of train data which we given the 33% for validation set for calculating the loss we need to give the validation set which means loss function to find the minimal distance between the actual target feature values  and predicted values and we get predicted values by this validation set , and batch size we given as 10 which means at a time 10 records are going to feed to model and get perform the F+B propagation to miinze the loss function by using opitimizer and we given max Forward+backward propagation for each batch size(10) as 1000  for updating the weights
```

As we can see above when the val_loss is not getting improve we are performing the early stoping the trainig of the mode so we can come to conclude that we can get to know how namy epoches we need to use by this early stopping 

### Details about 1 epoch completion
One Epoch:

- The entire training dataset (X_train) is divided into batches of 10 records each.
- For X_train with 5360 records and a batch size of 10, there will be 536 batches in one epoch.
- During each epoch, all 5360 records will be processed in 536 batches.
- After processing all batches (i.e., after 536 forward and backward propagations), one epoch is completed.

```python
# summarize history for accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
